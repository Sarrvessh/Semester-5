```python
def Bagging(n_estimator,data,list_test):
    Y_pred = []
    num_sample_instances = len(data)//n_estimator
    for i in range(n_estimator):
        df = data.sample(num_sample_instances)
        X = df.drop('play',axis=1)
        Y = df['play']
        clf = LogisticRegression()
        clf_model = clf.fit(X,Y)
        pred = clf_model.predict(list_test)
        Y_pred.append(pred)
    count_Y = Y_pred.count(1)
    count_N = Y_pred.count(0)
    if count_Y>count_N :
        print("The resultant prediction is Yes. The person enjoys")
    else :
        print("The resultant prediction is No. The person does not enjoy")
    Y_pred = []
```
